### Diabetes Status Subset
- (Ever told) (you had) diabetes
- create a full-population dataframe using only `diabete4`
- drop missing values for the diabetes-status variable


```python
diabetes_status_df = brfss_2021_df[['diabete4']].dropna()
diabetes_status_df.shape
```

### Clean Diabetes Status
- remove invalid `diabete4` codes for don't know and refused
- inspect the remaining diabetes-status counts


```python
diabetes_status_df = diabetes_status_df[diabetes_status_df['diabete4'] != 7]
diabetes_status_df = diabetes_status_df[diabetes_status_df['diabete4'] != 9]
diabetes_status_df['diabete4'].value_counts().sort_index()
```

### Recode Diabetes Target
- convert BRFSS diabetes codes into the `0/1/2` target used in the reference notebook


```python
diabetes_status_df['Diabetes_012'] = diabetes_status_df['diabete4'].replace({
    2: 0,
    4: 0,
    3: 1,
    1: 2,
})

diabetes_status_df.groupby(['Diabetes_012']).size()
```

### Feature List
- define the diabetes-management variables and covariates used in the paper-style analysis dataframe


```python
columns = [
"diabedu",
"feetchk3",
"bldsugar",
"doctdiab",
"chkhemo3",
"feetchk",
"eyeexam1",
"menthlth",
"educa",
"x.ageg5yr",
"birthsex",
"marital",
"employ1",
"income3",
"x.urbstat",
"persdoc3",
"x.hlthpln",
"addepev3",
"diabete4",
"diabage3"
]
```

### Select Analysis Columns
- build `features_df` from the selected diabetes-management variables


```python
features_df = brfss_2021_df[columns].copy()
```

### Analysis Shape

# Cleaning


### Drop Missing Values
- remove rows with missing values across the selected diabetes-management variables


```python
features_df = features_df.dropna()
features_df.shape
```

### diabedu
- whether respondent ever took a diabetes management class
- remove `7` and `9` because they represent don't know / refused
- change `2` to `0` so no is `0` and yes remains `1`


```python
features_df = features_df[features_df['diabedu'] != 7]
features_df = features_df[features_df['diabedu'] != 9]
features_df['diabedu'] = features_df['diabedu'].replace({2: 0})
features_df['diabedu'].unique()
```

### feetchk3
- how often respondent checks their feet for sores or irritation
- remove `777` and `999` because they represent don't know / refused
- keep the remaining numeric values so you can later convert this into daily vs less-than-daily if needed


```python
features_df = features_df[features_df['feetchk3'] != 777]
features_df = features_df[features_df['feetchk3'] != 999]
features_df['feetchk3'].unique()
```

### bldsugar
- how often respondent checks blood sugar
- remove `777` and `999` because they represent don't know / refused
- keep the remaining numeric values so you can later convert this into daily vs less-than-daily if needed


```python
features_df = features_df[features_df['bldsugar'] != 777]
features_df = features_df[features_df['bldsugar'] != 999]
features_df['bldsugar'].unique()
```

### doctdiab
- number of times respondent saw a health professional for diabetes in the past 12 months
- remove `77` and `99` because they represent don't know / refused
- keep the remaining counts as numeric values


```python
features_df = features_df[features_df['doctdiab'] != 77]
features_df = features_df[features_df['doctdiab'] != 99]
features_df['doctdiab'].unique()
```

### chkhemo3
- number of A1c checks in the past year
- remove `77` and `99` because they represent don't know / refused
- keep the remaining numeric counts so you can later convert this into at least twice per year vs less than twice


```python
features_df = features_df[features_df['chkhemo3'] != 77]
features_df = features_df[features_df['chkhemo3'] != 99]
features_df['chkhemo3'].unique()
```

### feetchk
- number of times respondent had their feet checked by a health professional in the past 12 months
- remove `77` and `99` because they represent don't know / refused
- keep the remaining values as valid counts so this can later be recoded into annual foot check vs not annual foot check


```python
features_df = features_df[features_df['feetchk'] != 77]
features_df = features_df[features_df['feetchk'] != 99]
features_df['feetchk'].unique()
```

### eyeexam1
- how recently respondent had a dilated eye exam
- remove `7` and `9` because they represent don't know / refused
- keep the remaining categories so you can later convert this into annual exam vs not annual


```python
features_df = features_df[features_df['eyeexam1'] != 7]
features_df = features_df[features_df['eyeexam1'] != 9]
features_df['eyeexam1'].unique()
```

### menthlth
- number of bad mental health days in the past 30 days
- change `88` to `0` because BRFSS uses `88` for none
- remove `77` and `99` because they indicate don't know / refused
- keep this as a count variable from `0` to `30`


```python
features_df['menthlth'] = features_df['menthlth'].replace({88: 0})
features_df = features_df[features_df['menthlth'] != 77]
features_df = features_df[features_df['menthlth'] != 99]
features_df['menthlth'].unique()
```

### educa
- respondent education level
- remove `9` because it represents refused / missing
- keep the remaining categories as an ordinal education variable


```python
features_df = features_df[features_df['educa'] != 9]
features_df['educa'].unique()
```

### x.ageg5yr
- respondent age group
- remove `14` because it corresponds to missing / refused rather than a real age group
- keep the remaining ordered age-group categories as valid responses


```python
features_df = features_df[features_df['x.ageg5yr'] != 14]
features_df['x.ageg5yr'].unique()
```

### birthsex
- whether respondent is male
- change `2` to `0` so female is `0` and male remains `1`
- this makes the variable binary and easier to interpret in modeling


```python
features_df['birthsex'] = features_df['birthsex'].replace({2: 0})
features_df['birthsex'].unique()
```

### marital
- respondent marital or partnered status
- remove `9` because it represents refused
- keep the remaining categories so you can later collapse them into married/partnered vs not married/partnered


```python
features_df = features_df[features_df['marital'] != 9]
features_df['marital'].unique()
```

### employ1
- respondent employment status
- remove `9` because it represents refused
- keep the remaining categories so you can later collapse them into employed vs not employed


```python
features_df = features_df[features_df['employ1'] != 9]
features_df['employ1'].unique()
```

### income3
- annual household income category
- remove `77` and `99` because they represent don't know / refused
- keep the remaining categories as an ordinal income variable


```python
features_df = features_df[features_df['income3'] != 77]
features_df = features_df[features_df['income3'] != 99]
features_df['income3'].unique()
```

### x.urbstat
- whether respondent lives in an urban or rural county
- inspect values here because this is a derived variable
- if only valid categories are present, no additional cleaning is needed


```python
features_df['x.urbstat'].unique()
```

### persdoc3
- whether respondent has at least one personal health care provider
- remove `7` and `9` because they represent don't know / refused
- keep the valid categories, then collapse later if you want at least one provider vs none


```python
features_df = features_df[features_df['persdoc3'] != 7]
features_df = features_df[features_df['persdoc3'] != 9]
features_df['persdoc3'].unique()
```

### x.hlthpln
- whether respondent has any health care coverage
- remove `7` and `9` because they represent don't know / refused
- change `2` to `0` so no coverage is `0` and yes remains `1`


```python
features_df = features_df[features_df['x.hlthpln'] != 7]
features_df = features_df[features_df['x.hlthpln'] != 9]
features_df['x.hlthpln'] = features_df['x.hlthpln'].replace({2: 0})
features_df['x.hlthpln'].unique()
```

### addepev3
- whether respondent was ever told they had a depressive disorder
- remove `7` and `9` because they represent don't know / refused
- change `2` to `0` so no is `0` and yes remains `1`


```python
features_df = features_df[features_df['addepev3'] != 7]
features_df = features_df[features_df['addepev3'] != 9]
features_df['addepev3'] = features_df['addepev3'].replace({2: 0})
features_df['addepev3'].unique()
```

### diabete4
- remove `7` and `9` because they represent don't know / refused
- inspect the remaining diabetes-status values in the diabetes-management subset
- keep this as an inspection step rather than forcing the dataset to diagnosed diabetes only


```python
features_df = features_df[features_df['diabete4'] != 7]
features_df = features_df[features_df['diabete4'] != 9]
features_df.groupby(['diabete4']).size()
```

### diabage3
- respondent age when first told they had diabetes
- remove `98` and `99` because they represent don't know / refused
- keep the remaining values as a numeric age-at-diagnosis variable


```python
features_df = features_df[features_df['diabage3'] != 98]
features_df = features_df[features_df['diabage3'] != 99]
features_df['diabage3'].unique()
```

### Rename Management Columns
- rename the cleaned diabetes-management variables to readable analysis names before EDA and export


```python
features_df = features_df.rename(columns={
    'diabedu': 'Diabetes_Class',
    'feetchk3': 'Foot_Self_Check_Freq',
    'bldsugar': 'Blood_Sugar_Check_Freq',
    'doctdiab': 'Doctor_Visit_Diabetes_Freq',
    'chkhemo3': 'A1C_Check_Freq',
    'feetchk': 'Professional_Foot_Check_Freq',
    'eyeexam1': 'Eye_Exam_Recency',
    'menthlth': 'Bad_Mental_Health_Days',
    'educa': 'Education',
    'x.ageg5yr': 'Age_Group',
    'birthsex': 'Male',
    'marital': 'Marital_Status',
    'employ1': 'Employment_Status',
    'income3': 'Income',
    'x.urbstat': 'Urban_Rural_Status',
    'persdoc3': 'Personal_Doctor',
    'x.hlthpln': 'Health_Coverage',
    'addepev3': 'Depressive_Disorder',
    'diabete4': 'Diabetes_Status',
    'diabage3': 'Diabetes_Age'
})
```
