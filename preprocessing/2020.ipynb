## Core 1: Health Status

### genhlth
**Question:** Would you say that in general your health is excellent, very good, good, fair, or poor?
- `1` = Excellent
- `2` = Very Good
- `3` = Good
- `4` = Fair
- `5` = Poor
- `7` = Don't know / Not sure
- `9` = Refused
- cleaning: set `7` and `9` to `pd.NA`


```python
df_copy['genhlth'] = df_copy['genhlth'].replace([7, 9], pd.NA)
df_copy['genhlth'].value_counts(dropna=False)
```
